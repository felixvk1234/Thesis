```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'M1': ["", "", ""],
    'M2': ["Social Network (M2)", "", ""],
    'M3': ["Labels (M3)", "Social Network (M3)", ""],
    'M4': ["", "Labels (M4)", "Social Network (M4)"],
    'M5': ["", "", "Labels (M5)"],
}

# Index
phases = ['Training', 'Validation', 'Testing']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Define the timeline
months = ["M1", "M2", "M3", "M4", "M5"]
x = np.arange(len(months) + 1)  # Define gridlines at full steps

# Plot horizontal bars for each phase
for i, phase in enumerate(phases):
    for j, month in enumerate(months):
        if data[month][i]:  # If there's data for this phase and month
            ax.barh(y=i, width=1, height=0.4, left=j, color=f'C{i}', edgecolor='black', alpha=0.7)
            ax.text(j + 0.5, i, data[month][i], ha='center', va='center', fontsize=10, color='black')

#  Set x-ticks at midpoints between gridlines
ax.set_xticks(x[:-1] + 0.5)  
ax.set_xticklabels(months, fontsize=12)

# Add phase labels on the y-axis
ax.set_yticks(np.arange(len(phases)))
ax.set_yticklabels(phases, fontsize=12)

# Add a title and labels
ax.set_title("Short-Term Model", fontsize=16, pad=20)

# Add vertical gridlines at integer x-positions
ax.set_xticks(x, minor=True)
ax.grid(axis="x", linestyle="--", alpha=0.7, which="minor")

# Make it visually appealing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Long-Term Network
data = {
    'M1': ["Social Network (M1-M2)", "", ""],
    'M2': ["Social Network (M1-M2)", "Social Network (M2-M3)", ""],
    'M3': ["Labels (M3)", "Social Network (M2-M3)", ""],
    'M4': ["", "Labels (M4)", "Social Network (M4)"],
    'M5': ["", "", "Labels (M5)"],
}

# Index
phases = ['Training', 'Validation', 'Testing']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Define the timeline
months = ["M1", "M2", "M3", "M4", "M5"]
x = np.arange(len(months) + 1)  # Define gridlines at full steps

# Plot horizontal bars for each phase
for i, phase in enumerate(phases):
    if phase == 'Training':
        # Training: Social Network (M1-M2)
        ax.barh(y=i, width=2, height=0.4, left=0, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(1.0, i, "Social Network (M1-M2)", ha='center', va='center', fontsize=10, color='black')
        # Training: Labels (M3)
        ax.barh(y=i, width=1, height=0.4, left=2, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(2.5, i, "Labels (M3)", ha='center', va='center', fontsize=10, color='black')
    elif phase == 'Validation':
        # Validation: Social Network (M2-M3)
        ax.barh(y=i, width=2, height=0.4, left=1, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(2.0, i, "Social Network (M2-M3)", ha='center', va='center', fontsize=10, color='black')
        # Validation: Labels (M4)
        ax.barh(y=i, width=1, height=0.4, left=3, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(3.5, i, "Labels (M4)", ha='center', va='center', fontsize=10, color='black')
    elif phase == 'Testing':
        # Testing: Social Network (M4)
        ax.barh(y=i, width=1, height=0.4, left=3, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(3.5, i, "Social Network (M4)", ha='center', va='center', fontsize=10, color='black')
        # Testing: Labels (M5)
        ax.barh(y=i, width=1, height=0.4, left=4, color=f'C{i}', edgecolor='black', alpha=0.7)
        ax.text(4.5, i, "Labels (M5)", ha='center', va='center', fontsize=10, color='black')

#  Set x-ticks at midpoints between gridlines
ax.set_xticks(x[:-1] + 0.5)
ax.set_xticklabels(months, fontsize=12)

# Add phase labels on the y-axis
ax.set_yticks(np.arange(len(phases)))
ax.set_yticklabels(phases, fontsize=12)

# Add a title and labels
ax.set_title("Long-Term Model", fontsize=16, pad=20)

#  Add minor gridlines at full integer x-values
ax.set_xticks(x, minor=True)
ax.grid(axis="x", linestyle="--", alpha=0.7, which="minor")

# Make it visually appealing
plt.tight_layout()

# Show the plot
plt.show()
```
